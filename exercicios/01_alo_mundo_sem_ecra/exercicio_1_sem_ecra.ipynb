# Exercício 1 — sem ecrã (chat com **histórico**, DeepSeek)

Código **só neste caderno** (sem `agent.py`): uma lista de mensagens LangChain (`HumanMessage` / `AIMessage`) e uma chamada `invoke` por turno.

**Jupyter no Docker:** nesta pasta, **`./run.sh`**. Variável **`DEEPSEEK_MODEL`** (predefinição típica: `deepseek-chat`).

Chave: **`DEEPSEEK_API_KEY`** no `.env` na raiz do repositório. Opcional: **`DEEPSEEK_API_BASE`** (predefinição: `https://api.deepseek.com`).

```python
import os

api_key = (os.environ.get("DEEPSEEK_API_KEY") or "").strip()
if not api_key:
    raise RuntimeError(
        "Defina DEEPSEEK_API_KEY no `.env` na raiz do repositório (exercício 1 — DeepSeek)."
    )

modelo = (os.environ.get("DEEPSEEK_MODEL") or "deepseek-chat").strip() or "deepseek-chat"
base_url = (os.environ.get("DEEPSEEK_API_BASE") or "https://api.deepseek.com").strip().rstrip("/")
print(f"Modelo: {modelo}", flush=True)
print(f"API base: {base_url}", flush=True)
```

```python
from langchain_core.messages import AIMessage, HumanMessage
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model=modelo,
    api_key=api_key,
    base_url=base_url,
    temperature=0.2,
)

mensagens = [HumanMessage(content="Explica numa frase o que é o LangChain.")]
r = llm.invoke(mensagens)
resposta = r.content if isinstance(r.content, str) else str(r.content)
print("Assistente:", resposta, flush=True)

mensagens.append(AIMessage(content=resposta))
mensagens.append(HumanMessage(content="Agora resume isso numa só palavra em português."))
r2 = llm.invoke(mensagens)
resposta2 = r2.content if isinstance(r2.content, str) else str(r2.content)
print("Assistente:", resposta2, flush=True)
```
